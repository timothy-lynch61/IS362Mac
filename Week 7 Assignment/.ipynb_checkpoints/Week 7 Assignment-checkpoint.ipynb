<h1><center> Load the ratings by user information that you collected into a pandas dataframe </center></h1>

```python
#Import pandas
import pandas as pd

#Import data and set columns to float
movie_data = pd.read_csv('movies.csv', dtype = {'American Sniper':float, 'Edge of Tomorrow':float, 'Groundhog Day':float, 'Jurassic World':float, 'Lost in Translation':float, 'Lucy':float})

#Import Sklearn to help with normalization
from sklearn import preprocessing

#Look at the data
movie_data
```

<h1>Show the average ratings for each user and each movie</h1>

Thanks to the skipna parameter we can find the mean without having to change the NaN values. This results in a truer mean than just converting the NaN's to 0 and then using .mean()

```python
#Next we are going to show the average rating per user
movie_data['User Average'] = movie_data.mean(axis = 1, skipna = True)

#Then the average per movie
movie_data.loc['Average Per Movie'] = movie_data.mean(axis = 0, skipna = True)
movie_data
```

```python
#Clean up some of the data
movie_data.loc['Average Per Movie', 'User Average'] = None
movie_data.loc['Average Per Movie', 'Names'] = None
movie_data
```

<h1><center> Create a new pandas dataframe, with normalized ratings for each user. Again, show the average ratings for each user and each movie.</center></h1>

```python
#Normalizing the Movies

normalized_data = movie_data.drop('Names', axis = 1)


# This is where things get tricky, I'm not solid in statistics

min_max_scaler = preprocessing.MinMaxScaler()
scaled = min_max_scaler.fit_transform(normalized_data)
movie_normalized = pd.DataFrame(scaled)

# Create Columns
column_names = ['American Sniper', 'Edge of Tomorrow', 'Groundhog Day', 'Jurassic World', 'Lost in Translation', 'Lucy', 'User Average']
movie_normalized.columns = column_names
movie_normalized
```
