```python
#Import Pandas
import pandas as pd

#Load in Data, please note that encoding had to be changed for the data to work

ndc = pd.read_csv('product.csv', encoding = "ISO-8859-1")

#Loading a little bit of the data

ndc.head(4)
```

__Transforming the data__ 

Before I start working on the data, I would like to actually copy it. This way I can compare the two later on if need be. The next step I would like to do is make the column names lower case. This is to help eliminate programming mistakes. After that we are going to remove NaN values. If the column is a __string__ we are going to replace it with a "Not Entered" string. If the value is an integer or a float we will use the number 0 to replace the NaN value.

__First Things First__

```python
#Copy the data
ndc2 = ndc.copy()

#Convert upper case columns to lower
ndc2.columns = ndc2.columns.str.lower()

#Example of column change
ndc2.head(4)
```

__Finding NaN Values__

Next we are going to check the columns with NaN value for their types. Going off of what we can see below that is columns "proprietarynamesuffix, endmarketingdate,deaschedule and pharm_classes". I have already done some research on that and know that all of them are strings.Since they are all strings we are going to replace them with the values "Not Entered". 

```python
# Will need to run these on their own cell, but all of them are stings

type("proprietarynamesuffix")
type("endmarketingdate")
type("deaschedule")
type("pharm_classes")
```

```python
# Converting NaN to "Not Entered"

ndc2['proprietarynamesuffix'].fillna(value = 'Not Entered', inplace = True)
ndc2['endmarketingdate'].fillna(value = 'Not Entered', inplace = True)
ndc2['deaschedule'].fillna(value = 'Not Entered', inplace = True)
ndc2['pharm_classes'].fillna(value = 'Not Entered', inplace = True)

#Inspect the NaN Value change
ndc2.head(4)
```

# Second Data Set Added

__eDPI settings for competitive video gaming__

For this dataset we want to know the average eDPI for one of the teams. Since we want to know about teams, we can clean up the data by dropping every row that doesn't have a Team name or a eDPI score. 



```python
# Importing and loading data

import pandas as pd
gaming = pd.read_csv('CS_GO.csv')
gaming.head()
```

__Cleaning up Data__

As you can see below a lot of the values in the csv file have NaN's. Since the only thing we are interested in are the Team name and eDPI values we will only be cleaning those up this go around. I did change the name of column "Team name" to "Team_name" for programming purposes

Since we want to know the average eDPI for a specific team we will drop any of the rows that are missing values for Team name or eDPI. 

```python
# Cleaning up Data

#Drops Rows if either Team name or eDPI have values of NaN
gaming.dropna(subset=['Team name', 'eDPI'], how = 'any', inplace = True)
gaming.head(5)
```

```python
#Changing Team name to Teamname
gaming.rename (columns = {'Team name': 'Team_name'}, inplace = True)

gaming.head(5)
```

__Calculate the Result__

Since we know have our data as clean as it needs to be (we could of done more), we can now work with it. The question asks us to find the average eDPI of a particular team. For us we are going to search for the mean of the team __Natus Vincere__.

As we can see below the average for eDPI for team __Natus Vincere__ is 1007.2 



```python
average_ePDI = gaming.loc[gaming.Team_name == 'Natus Vincere', ['Team_name', 'eDPI']]
```

```python
average_ePDI.mean()
```
