# The first Question is: What is the northernmost airport in the United State

The correct answer is Wiley Post Will Rogers Memory Airport. You will notice that it is second on the list. The first one Dillant Hopkins Airport is actually incorrect. I found this out by using the .max() method to find the highest latitude. Then I wanted to know if that number was right, I took a guess and figured that if I filtered by latitude near 70 degrees I would get a few more airports. 

I then took the list and googled Dillant Hopkins Airport. It actually is not the northern most airport. The lat and lon coordinates were put in incorrectly. It comes no where close with a latitude that is supposed to be 42 degrees. 

```python
airplanes['lat'].max()
```

```python
air2 = airplanes.loc[airplanes.lat > 70, ['name','tzone','lat','lon']]
```

```python
air2.sort_values('lat',ascending = False)
```

# The second Question is: What is the easternmost airport in the United States

# The third Question is: On Feb 12th, 2013, which New York Area airport had the windiest weather

For this question we need to import the Weather CSV. After that, below I filter out the day and month to get the data we want. And then I sort by wind_speed. According to the data EWR had the highest wind speed of 1048. I believe that data to be outside the usual realm, so I would pick the runner up which is LGA which had a wind speed of 23 and gust of 31 to be the most accurate. 

```python
#Import CSV of Weather Data
weather = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv')
```

```python
weather.head(10)
```

```python
weather2 = weather[(weather['month'] == 2) & (weather.day == 12)]
```

```python
weather2.sort_values('wind_speed', ascending = False).head(5)
```
